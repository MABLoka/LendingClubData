```python
import pandas as pd

df = pd.read_csv('../data/loan_raw.csv')
df.shape
df.head()
```

```python
df.info()
df.describe()
```

```python
missing = df.isnull().sum()
missing_percent = (missing / len(df)) * 100
missing_df = pd.DataFrame({'Missing': missing, 'Percent': missing_percent})
missing_df[missing_df['Missing'] > 0].sort_values('Percent', ascending=False)
```

```python
unique_counts = df.nunique().sort_values()
unique_counts.head(10)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Income
threshold = df['annual_inc'].quantile(0.99)
filtered = df[df['annual_inc'] <= threshold]

filtered['annual_inc'].hist(bins=50, figsize=(8, 6))
plt.title("Annual Income (Capped at 99th Percentile)")
plt.xlabel("Annual Income")
plt.ylabel("Frequency")
plt.show()


# Loan amount
df['loan_amnt'].describe()

# Grade
df['grade'].value_counts(normalize=True)

# Loan status
df['loan_status'].value_counts()
```
